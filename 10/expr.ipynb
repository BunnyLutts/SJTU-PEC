```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import linregress

d1 = np.array([50.39320, 50.40870, 50.42470, 50.43960, 50.45455, 50.46940, 50.48410])
n = np.arange(0, len(d1))

slope, intercept, r_value, p_value, std_err = linregress(n, d1)

plt.plot(d1, 'x', label='Measured')
plt.plot(n, slope*n + intercept, '-', label=f'Fitted: y = {slope}x + {intercept}, $r^2$ = {r_value**2}')
plt.xlabel("n")
plt.ylabel("$d_1$ (mm)")
plt.legend()
```

```python
lbd = 2*slope/50
print("lambda =", lbd)
```

```python
d2 = np.array([50.565, 50.855, 51.155, 51.735])
n = np.arange(0, len(d2))
slope, intercept, r_value, p_value, std_err = linregress(n, d2)

plt.plot(d2, 'x', label='Measured')
plt.plot(n, slope*n + intercept, '-', label=f'Fitted: y = {slope}x + {intercept}, $r^2$ = {r_value**2}')
plt.xlabel("n")
plt.ylabel("$d_2$ (mm)")
plt.legend()
```

```python
dlbd = lbd**2 / (2*slope)
print("dlambda =", dlbd)
```
