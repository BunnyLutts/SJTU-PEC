```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('data.csv')
U = data['U']
f = data['f']
dt = data['dt']
f0 = f[0]
Um = U[0]

indeces = f.argsort()
f = f[indeces]
U = U[indeces]
dt = dt[indeces]

df = 5.953-4.244
print(f"df={df}, Q={f0/df}")

plt.plot(f/f0, U/Um)
plt.xlabel('f/f0')
plt.ylabel('U/Um')
plt.show()
```

```python
phi = dt*10**-6 *2*np.pi*f*10**3
plt.plot(f/f[0], phi, '*', label='phase (Measured)')
plt.plot(f/f[0], np.full_like(f/f[0], 0), 'b--')
plt.plot(f/f[0], np.sign(f-f[0])*np.arccos(U/Um), label="phase' (Calculated)")
plt.xlabel('f/f0')
plt.ylabel('phase (rad)')
plt.legend()
plt.show()
```
