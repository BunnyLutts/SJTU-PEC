```python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import linregress

L1 = np.array([74.00, 78.63, 83.19, 87.81, 92.32, 96.86, 101.48, 105.92, 110.68, 114.98])
L2 = np.array([78.68, 88.03, 97.66, 106.30, 115.46, 124.54, 133.74, 142.81, 151.92, 160.98])
L3 = np.array([100.00, 110.00, 120.00, 130.00, 140.00, 150.00, 160.00])
t3 = np.array([86.40, 92.40, 99.20, 106.0, 112.4, 119.2, 126.0])

def plotLine(x, y, labelx, labely, label):
    plt.plot(x, y, 'x', label='original')
    result = linregress(x, y)
    plt.plot(x, result.slope*x + result.intercept, label='fitted')
    plt.text(x[-1], y[0], f'$y = {result.slope:.4f}x + {result.intercept:.4f}$\n$R$ = {result.rvalue:.4f}', horizontalalignment='right', verticalalignment='bottom')
    plt.xlabel(labelx)
    plt.ylabel(labely)
    plt.legend()
    plt.title(label)
    plt.show()
    return result
```

```python
plotLine(np.arange(1, len(L1)+1), L1, 'i', 'L(mm)', '')
```

```python
plotLine(np.arange(1, len(L2)+1), L2, 'i', 'L(mm)', '')
```

```python
plotLine(t3, L3, 't($\mu s$)', 'L(mm)', '')
```
